# Estimate the variance of gradients to test the barren plateau

```python
import scipy.integrate as integrate
import numpy as np
```

```python
# Define the integrand
def var_of_grad(n, gamma=-20, a=-np.pi, b=np.pi):
    # Integration bounds (0.2 to 0.4 for each variable)
    bound = [(a, b)]
    bounds = bound * n

    def integrand(theta_k, *THETAL):
        b = 1
        for theta_l in THETAL:
            b *= np.cos(theta_l / 2) ** 2

        f_theta_k = np.sin(theta_k) ** 2 / ((np.exp(gamma) + (1- np.exp(gamma)) * (np.cos(theta_k / 2) ** 2) * b) ** 2)
        f_theta = (((np.exp(gamma) -1) / (2 * gamma)) ** 2) * (b ** 2) * f_theta_k
        return f_theta

    # Perform the n-dimensional integration
    result, _ = integrate.nquad(integrand, bounds)
    result = result / ((2 * np.pi) ** n)
    return result

var_of_grad(2)
```
